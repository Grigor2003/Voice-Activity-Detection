```python
import os
import shutil
import pandas as pd
```

```python
data_dir = 'accent-dataset/common_accent'
new_class = 'other'
new_class_dir_path = os.path.join(data_dir, new_class)
os.makedirs(new_class_dir_path, exist_ok=True)
with open(os.path.join(new_class_dir_path, 'accent.txt'), 'w') as f:
    f.write(new_class)
old_classes = [
    'filipino',
    'african',
    'hong_kong',
    'singaporean'
]
old_labels_path = 'accent-dataset/16000_silero_labels.csv'
new_labels_path = 'accent-dataset/16000_silero_labels_concated.csv'

old_labels = pd.read_csv(old_labels_path)
file_label_dict = dict(zip(old_labels['filename'], old_labels['labels']))
new_file_label_dict = dict()
```

```python
file_label_dict
```

```python
for c in old_classes:
    old_class_dir = os.path.join(data_dir, c)
    for file in os.listdir(old_class_dir):
        if not file.endswith('.wav'): continue
        old_path = os.path.join(old_class_dir, file)
        old_label_key = os.path.join(c, file)
        label = file_label_dict[old_label_key]
        
        new_filename = os.path.join(new_class, f'{c}_{file}')
        new_path = os.path.join(data_dir, new_filename)
        print(new_path)
        os.rename(old_path, new_path)
        new_file_label_dict[new_filename] = label
        file_label_dict.pop(old_label_key)
    shutil.rmtree(old_class_dir)
```

```python
new_file_label_dict.update(file_label_dict)
```

```python
with open(new_labels_path, 'w') as f:
    f.write('filename,labels\n')
    f.write('\n'.join(f'{k},{v}' for k,v in new_file_label_dict.items()) + '\n')
```
